```python
#Q1. Read the dataset to python environment.
iris_data = pd.read_excel(r"C:\Users\USER\Desktop\ICT DATASCIENCE\iris.xls")
```

```python
iris_data
```

```python
#Q2. Display the columns in the dataset.
print("Columns in the dataset:")
print(iris_data.columns)
```

```python
#Q3. Calculate the mean of each column of the dataset.
means = iris_data.mean(numeric_only=True)
print("\nMean of each column:")
print(means)
```

```python
#Q4. Check for the null values present in the dataset.
null_val = iris_data.isnull().sum()
print("\nNull values in the dataset:")
print(null_val)
```

```python
#Q5. Perform meaningful visualizations using the dataset.
#Bring at least 3 visualizations.
#V1: Boxplot for SL by SW
sns.boxplot(x='SW', y='SL', data=iris_data)
plt.title('Boxplot of SL by SW')
plt.show()
```

```python
#V2: Bar chart for PW 
sns.barplot(x=iris_data['Classification'], y='PW', data=iris_data)
plt.title('Bar Chart for PW by Species')
plt.show()
```

```python
#V3 :Scatter plot for SL and SW
sns.scatterplot(x='SL', y='SW', hue='Classification', data=iris_data)
plt.title('Scatter Plot for SL and SW')
plt.show()
```
